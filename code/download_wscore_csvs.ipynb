```python
import flywheel
import os
import datetime
from dateutil.tz import tzutc

fw = flywheel.Client()
gear = fw.lookup('gears/wscore-ct-heatmap/0.2.1_0.2.0')
collection_id = '6184053063bd671113926cc7'

local_dir = '../wscore_data'
os.makedirs(local_dir, exist_ok=True)
```

```python
# Get all the sessions within the collection.
sessions = [fw.get_session(x.id) for x in fw.get_collection_sessions(collection_id)]
```

```python
for session in sessions:
    sub_label = fw.get(session.parents['subject']).label.replace('_','x')

    for analysis in session.analyses:
        if (analysis.label.startswith(('wscore_2021-05-26', 'wscore_2021-06-11','wscore_2021-11-04')) and analysis.job['state']=='complete' and analysis.files:
            file = [f for f in analysis.files if f.name.endswith('wscores.csv')]
            try:
                print(file[0].name)
                analysis.download_file(file[0].name, os.path.join(local_dir, file[0].name))
            except IndexError as e:
                print(session.label)
```
