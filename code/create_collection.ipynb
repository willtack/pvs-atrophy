```python
import flywheel
import pandas as pd

fw = flywheel.Client()
collection_id = '6184053063bd671113926cc7' # already created blank on Flywheel GUI
project_id = '5c508d5fc2a4ad002d7628d8' # NACC
project = fw.get(project_id)
```

```python
df = pd.read_csv('../PVS_subjects.csv')
df = df.dropna()
```

```python
session_ids = []
session_labels = []
for index, row in df.iterrows():
    inddid = str(int(row.INDDID))
    date = row.MRIDate.split('/')
    ses_label = "%s.*x%s%s%sx3T" % (inddid, date[2],date[0], date[1])
    session = project.sessions.find(f"label=~{ses_label}")[0]
    print("%s %s" % (session.label, session.id))
    session_ids.append(session.id)
    session_labels.append(session.label)
```

```python
df.insert(2,"SessionLabel", session_labels, False)
df
```
