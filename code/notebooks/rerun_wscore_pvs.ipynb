```python
import flywheel
import pandas as pd
fw = flywheel.Client()
gear = fw.lookup('gears/wscore-ct-heatmap/0.2.1_0.2.0')
collection_id = '60a4032befa949862afeafbb'
```

```python
sessions = [fw.get_session(x.id) for x in fw.get_collection_sessions(collection_id)]
```

```python
df = pd.read_excel('../../PVS_data.xlsx')
```

```python
for session in sessions:
    match=False
    for analysis in session.analyses:
        if 'WRONGAGE' in analysis.label:
            match=True
            print("MATCH")
            print(session.label)
            
    for analysis in session.analyses:          
        if 'antsct_2021-05-13_WT_PVS' in analysis.label and analysis.job['state']=='complete' and analysis.files and match:
            # get antsct zip file
            zip_file = [f for f in analysis.files if f.name.endswith('.zip')][0]
            print(zip_file.name)
            # get patient age
            sub = fw.get(session.parents['subject'])
            age_years = int(df.loc[df['SessionID']==session.label].AgeAtMRI)
            print(age_years)
            
            inputs = {"antsct_output": zip_file}
            config = {"patient_age": age_years, 'wthresholds': "0.0 0.5 1.0"}
            analysis_id = gear.run(analysis_label='wscore_2021-06-11_WT', config=config, inputs=inputs, destination=session)
            
        else:
            continue 
    
```
